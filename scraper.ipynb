```python
import requests
from bs4 import BeautifulSoup
import time

url = "https://www.livecoinwatch.com/"

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3',
    'Referer': 'https://www.livecoinwatch.com/',  # Set the Referer header to the URL of the website
    'Accept-Language': 'en-US,en;q=0.9',
    'Connection': 'keep-alive',
}

```

```python
response = requests.get(url,headers=headers)
response
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
# Find all rows with class "table-row filter-row"
rows = soup.select('tr.table-row.filter-row')
```

```python
# Initialize empty lists to store data
coin_names = []
values_list = []

# Iterate over each row
for row in rows:
    # Find anchor element with class "text-left" inside a td with class "filter-item table-item"
    coin_link = row.select_one('td.filter-item.table-item a.text-left')
    
    # Continue to the next iteration if coin_link is None
    if coin_link is None:
        continue
    
    # Find all td elements with class "filter-item table-item"
    td_elements = row.select('td.filter-item.table-item')
    
    # Extract the coin name from the href attribute
    coin_name = coin_link.get_text(strip=True)
    
    # Extract other values from each td element
    values = [td.get_text(strip=True) for td in td_elements]
    
    # Append data to lists
    coin_names.append(coin_name)
    values_list.append(values)
```

```python
import pandas as pd
# Display the columns and values
columns = ['Index','Coin Name', 'Price', 'Market Cap', 'Volume 24H', 'Liquidity', 'All Time High', '1 H', '24 H',"asd"]
df = pd.DataFrame(values_list, columns=columns)
```
